```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Country Name': [
        'China', 'India', 'United States', 'Indonesia', 'Pakistan',
        'Nigeria', 'Brazil', 'Bangladesh', 'Russia', 'Mexico'
    ],
    '2020': [
        1402112000, 1380004000, 331002647, 273523621, 220892331,
        206139589, 212559417, 164689383, 145934460, 128932753
    ],
    '2021': [
        1409517000, 1393409000, 331893745, 274021604, 227015247,
        215073511, 213993437, 168287511, 145734038, 127792286
    ],
    '2022': [
        1412175000, 1406631000, 331893745, 275501339, 240485658,
        223804632, 215313498, 173989416, 144713314, 126705138
    ]
}

```

```python
df = pd.DataFrame(data)
df = df.sort_values(by='2022', ascending=False)

```

```python

plt.figure(figsize=(10, 5))
plt.bar(df['Country Name'], df['2022'], color='teal')
plt.title('Total Population by Country (2022)')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y')
plt.show()
```

```python

plt.figure(figsize=(8, 8))
plt.pie(df['2022'], labels=df['Country Name'], autopct='%1.1f%%', startangle=140)
plt.title('Population Share by Country (2022)')
plt.axis('equal')
plt.show()


```

```python

plt.figure(figsize=(10, 6))
plt.barh(df['Country Name'], df['2022'], color='darkcyan')
plt.title('Population by Country (2022)')
plt.xlabel('Population')
plt.ylabel('Country')
plt.tight_layout()
plt.grid(axis='x')
plt.show()
```

```python
countries_to_plot = ['India', 'China', 'United States']
df_line = df[df['Country Name'].isin(countries_to_plot)].set_index('Country Name').T

```

```python
plt.figure(figsize=(10, 5))
for country in df_line.columns:
    plt.plot(df_line.index, df_line[country], marker='o', label=country)

plt.title('Population Trend (2020–2022)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
